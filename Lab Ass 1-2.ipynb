```python
# Create a DataFrame
df = pd.DataFrame(data)
# Step 2: Clean and convert salary to integers
df["Salary"] = df["Salary (Rs./month)"].str.replace(",", "").astype(int)
```

```python
# Step 3: Normalize Age and Salary using Min-Max Normalization
scaler = MinMaxScaler()
df[["Age_normalized", "Salary_normalized"]] = scaler.fit_transform(df[["Age", "Salary"]])
```

```python
# Step 4: Visualization - Histogram of Age and Salary
plt.figure(figsize=(12, 5))
# Age histogram
plt.subplot(1, 2, 1)
sns.histplot(df["Age"], kde=True, bins=8, color="skyblue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Employees")
```

```python
# Salary histogram
plt.subplot(1, 2, 2)
sns.histplot(df["Salary"], kde=True, bins=8, color="salmon")
plt.title("Salary Distribution")
plt.xlabel("Salary ()")
plt.ylabel("Number of Employees")

plt.tight_layout()
plt.show()
```
