```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 500)

# 1. Sigmoid and its derivative
def sig(x):
    return 1 / (1 + np.exp(-x))
def sig_der(x):
    s = sig(x)
    return s * (1 - s)
plt.figure(figsize=(12, 8))
plt.plot(x, sig(x), label='Sigmoid', color='blue')
plt.plot(x, sig_der(x), label='Sigmoid Derivative', color='purple')
plt.title("Sigmoid Function and Its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 2. Tanh and its derivative
def tanh(x):
    return np.tanh(x)
def tanh_der(x):
    return 1 - np.tanh(x)**2
plt.plot(x, tanh(x), label='Tanh', color='blue')
plt.plot(x, tanh_der(x), label='Tanh Derivative', color='purple')
plt.title("Tanh Function and Its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 3. ReLU and its derivative
def relu(x):
    return np.maximum(0, x)
def relu_der(x):
    return np.where(x > 0, 1, np.where(x < 0, 0, 0.5))
plt.plot(x, relu(x), label='ReLU', color='blue')
plt.plot(x, relu_der(x), label='ReLU Derivative', color='purple')
plt.title("ReLU Function and Its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 4. Leaky ReLU and its derivative
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
def leaky_relu_der(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)
plt.plot(x, leaky_relu(x), label='Leaky ReLU', color='blue')
plt.plot(x, leaky_relu_der(x), label='Leaky ReLU Derivative', color='purple')
plt.title("Leaky ReLU Function and Its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 5. y = e^(-x)
y1 = np.exp(-x)
dy1 = -np.exp(-x)
plt.plot(x, y1, label='y = e^(-x)', color='blue')
plt.plot(x, dy1, label="y' = -e^(-x)", color='purple')
plt.title("y = e^(-x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 6. y = e^(-|x|)
y2 = np.exp(-np.abs(x))
dy2 = -np.exp(-np.abs(x)) * np.sign(x)
plt.plot(x, y2, label='y = e^(-|x|)', color='blue')
plt.plot(x, dy2, label="y' = -e^(-|x|) * sign(x)", color='purple')
plt.title("y = e^(-|x|)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 7. y = e^(-x^2)
y3 = np.exp(-x**2)
dy3 = -2 * x * np.exp(-x**2)
plt.plot(x, y3, label='y = e^(-x)', color='blue')
plt.plot(x, dy3, label="y' = -2x * e^(-x)", color='purple')
plt.title("y = e^(-x)")
plt.legend()
plt.grid(True)
plt.show()
```
