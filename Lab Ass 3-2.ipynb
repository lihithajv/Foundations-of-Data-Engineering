```python
import numpy as np
import matplotlib.pyplot as plt
from scipy . stats import norm
```

```python
N = 1000000 # number of samples
# ==== Case I : X1 , X2 ~ U (0 ,1) ====
X1 = np . random . uniform (0 ,1 , N )
X2 = np . random . uniform (0 ,1 , N )
Z1 = X1 + X2
```

```python
# ==== Case II : X1 , X2 ~ Normal (0 ,1) ====
X1 = np . random . normal (0 ,1 , N )
X2 = np . random . normal (0 ,1 , N )
Z2 = X1 + X2
```

```python
# ==== Case III : X1 ~ U (0 ,1) , X2 ~ U (0 ,2) ====
X1 = np . random . uniform (0 ,1 , N )
X2 = np . random . uniform (0 ,2 , N )
Z3 = X1 + X2
```

```python
# ==== Plot results ====
fig , axs = plt . subplots (3 , 1 , figsize =(10 , 12) )
# Case I
axs [0]. hist ( Z1 , bins =200 , density = True , alpha =0.6 , color = 'b' , label = "Simulated " )
axs [0]. set_title ( " Case I : Z = X1 + X2 , X1 , X2 ~ U (0 ,1) ( I r w i n Hall dist ) ")
axs [0]. set_xlim ([0 ,2])
axs [0]. grid ( True )
axs [0]. legend ()
# Case II
axs [1]. hist ( Z2 , bins =200 , density = True , alpha =0.6 , color = 'r' , label = "Simulated " )
x = np . linspace ( -6 ,6 ,500)
axs [1]. plot (x , norm . pdf (x , loc =0 , scale = np . sqrt (2) ) , 'k' , lw =2 , label = "Theoretical N (0 ,2) " )
axs [1]. set_title ( " Case II : Z = X1 + X2 , X1 , X2 ~ N (0 ,1) " )
axs [1]. grid ( True )
axs [1]. legend ()
# Case III
axs [2]. hist ( Z3 , bins =200 , density = True , alpha =0.6 , color = 'g' , label = "Simulated " )
axs [2]. set_title ( " Case III : Z = X1 + X2 , X1 ~ U (0 ,1) , X2 ~ U (0 ,2) ( Trapezoidal) " )
axs [2]. set_xlim ([0 ,3])
axs [2]. grid ( True )
axs [2]. legend ()
plt . tight_layout ()
plt . show ()
```
