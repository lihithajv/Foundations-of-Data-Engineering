```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, expon
```

```python
# Setup figure
fig, axs = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle("PDF, CDF, and PPF of Normal, Uniform, and Exponential Distributions", fontsize=16)
# ==== NORMAL DISTRIBUTION ====
x = np.linspace(-4, 4, 500)
p = np.linspace(0.001, 0.999, 500)  # for PPF
mu, sigma = 0, 1
axs[0,0].plot(x, norm.pdf(x, mu, sigma), 'b')
axs[0,0].set_title("Normal PDF (=0, =1)")

axs[0,1].plot(x, norm.cdf(x, mu, sigma), 'r')
axs[0,1].set_title("Normal CDF (=0, =1)")

axs[0,2].plot(p, norm.ppf(p, mu, sigma), 'g')
axs[0,2].set_title("Normal PPF (=0, =1)")
# ==== UNIFORM DISTRIBUTION ====
x = np.linspace(-2, 4, 500)
a, b = 0, 2
axs[1,0].plot(x, uniform.pdf(x, a, b), 'b')
axs[1,0].set_title("Uniform PDF (0,2)")

axs[1,1].plot(x, uniform.cdf(x, a, b), 'r')
axs[1,1].set_title("Uniform CDF (0,2)")

axs[1,2].plot(p, uniform.ppf(p, a, b), 'g')
axs[1,2].set_title("Uniform PPF (0,2)")
# ==== EXPONENTIAL DISTRIBUTION ====
x = np.linspace(0, 6, 500)
lam = 1.0
axs[2,0].plot(x, expon.pdf(x, scale=1/lam), 'b')
axs[2,0].set_title("Exponential PDF (lambda=1)")

axs[2,1].plot(x, expon.cdf(x, scale=1/lam), 'r')
axs[2,1].set_title("Exponential CDF (lambda=1)")

axs[2,2].plot(p, expon.ppf(p, scale=1/lam), 'g')
axs[2,2].set_title("Exponential PPF (lambda=1)")
# Beautify plots
for row in axs:
    for ax in row:
        ax.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
